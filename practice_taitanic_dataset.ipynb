The error in the previous cell was due to using `||` instead of `|` for the logical OR operator in the pandas filtering condition.

Here is the corrected code:

```python
#print data which age bellow 10 years and they are male

taitanic_df[(taitanic_df['Age'] < 10) | (taitanic_df['Sex'] == 'male')] ['Age']
```
